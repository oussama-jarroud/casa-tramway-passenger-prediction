Import des librairies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Chargement des données

```python
# Adapte le chemin et le nom du fichier à tes données
df = pd.read_csv('../data/raw/passengers_casatramway_raw.csv')
```

Aperçu initial des données

```python
print("Aperçu des 5 premières lignes :")
print(df.head())

print("\nInformations sur les types de données et les valeurs manquantes :")
print(df.info())

print("\nStatistiques descriptives :")
print(df.describe())
```

Vérification des valeurs manquantes

```python
print("\nNombre de valeurs manquantes par colonne :")
print(df.isnull().sum())
```

Premières visualisations

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Nb_Passagers'], kde=True) # Adapte le nom de la colonne
plt.title('Distribution du Nombre de Passagers')
plt.xlabel('Nombre de Passagers')
plt.ylabel('Fréquence')
plt.show()
```
